# Logistic Regression - Binary Classifier

## The Intuition

**Logistic Regression** $$ P(Y=1) = \frac{e^{\beta_0 + \beta_1X_1 + \beta_2X_2 + .. + \beta_mX_m}}{1 + e^{\beta_0 + \beta_1X_1 + \beta_2X_2 + .. + \beta_mX_m}}$$

$$ Odds = \frac{P(Y=1)}{1-P(Y=1)} = \frac{P(Y=1)}{P(Y=0)} = e^{\beta_0 + \beta_1X_1 + \beta_2X_2 + .. + \beta_mX_m} $$

$$ln\Bigg(\frac{P(Y=1)}{P(Y=0)}\Bigg) = \beta_0 + \beta_1X_1 + \beta_2X_2 + .. + \beta_mX_m$$

**Linear Regression** $$ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + .. + \beta_mX_m$$

In logistic regression, however the $X_i$ independent variables are dummy variables ie. they take the value of 0 or 1.

Consider, observations $X_1$ = 0 or 1

$$ ln\Bigg(\frac{P(Y=1|X_1 = 1)}{P(Y=0|X_1 = 1)}\Bigg) - ln\Bigg(\frac{P(Y=1|X_1 = 0)}{P(Y=0|X_1 = 0)}\Bigg) = (\beta_0 + \beta_1(1) + .. + \beta_mX_m) - (\beta_0 + \beta_1(0) + \beta_2X_2 + .. + \beta_mX_m) = \beta_1$$

$$ ln(odds(Y=1|X_1=1)) - ln(odds(Y=1|X_1=0)) = ln\Bigg(\frac{odds(Y=1|X_1=1)}{odds(Y=1|X_1=0)}\Bigg) = \beta_1 $$


## The Mathematical Derivation
The Logistic Classification model is implemented to classify a dataset $\mathcal{D}\{\mathcal{X}, \mathcal{Y}\}$ consisting of **$M$ datapoints (no. of people)** . We denote the **$i$th datapoint/individual** within the our dataset to be $\{x^{(i)}, y^{(i)}\}$, where $\{x^{(i)}\in \mathbb{R}^N\}$, where **$N$ is the number of datapoint features**. The binary label predicted by the classifier is refered to as $y^{(i)}\in \{0,1\}$. 


If we were to incorporate a bias within our model - the inputs need to be augmented with a fixed unit input $\hat{x}^{(i)}=(1, x^{(i)})$. Therefore our representation of the Classification model can be denoted by the below equation.

\begin{equation}
    \hat{y}^{(i)} =  \beta^T\hat{x}^{(i)} = \beta_0 + \sum_{n=0}^N \beta_n x_n^{(i)}
    \label{eq:model}
\end{equation}

In equation, $\beta$ is the model parameter vector, $\hat{y}^{(i)}$ the model predication for the $i$th datapoint. The probability of the dataset is therefore a product of Bernoulli distributions, with $\sigma$ being a sigmoid function:

$$ \sigma(x) = \frac{1}{1+e^{-x}}$$

\begin{equation}
\mathbb{P}(  y^{(i)} = 1 |  \hat{x}^{(i)}  ) = \sigma(\beta^T\hat{x}^{(i)})
\quad \quad 
\mathbb{P}(  y^{(i)} = 0 |  \hat{x}^{(i)}  ) = 1 - \sigma(\beta^T\hat{x}^{(i)})
\end{equation}

\begin{eqnarray}
    \mathbb{P}(\mathcal{Y}|\mathcal{X}, \beta) &=& \prod_{i=1}^{M}\mathbb{P}(  y^{(i)}  |  \hat{x}^{(i)}  )\\
    &=&  \prod_{i=1}^{M} \sigma(\beta^T\hat{x}^{(i)})^{y^{(i)}}(1-\sigma(\beta^T\hat{x}^{(i)}))^{1-y^{(i)}}
\end{eqnarray}

The gradients of the log-likelihood of the parameters $\frac{\partial }{\partial \beta} \mathcal{L}(\beta)$ where $\mathcal{L}(\beta) = \log \mathbb{P}(\mathcal{Y}|\mathcal{X}, \beta)$ are derived. Letting $u = \beta^T\hat{x}$:
\begin{eqnarray}
    Cost/Loss Function = \mathcal{L}(\beta) &=& \log \mathbb{P}(\mathcal{Y}|\mathcal{X}, \beta) \\
    &=& \sum_{i=1}^M y^{(i)}\log \sigma (u) + (1-y^{(i)})\log (1- \sigma (u))
\end{eqnarray}
Taking the derivative with respect to a single parameter $\beta_j$, $\frac{\partial }{\partial \beta_j}$ is:
\begin{eqnarray}
   \sum_{i=1}^M y^{(i)} \frac{\partial }{\partial \beta_j} \log \sigma (u) + (1-y^{(i)}) \frac{\partial }{\partial \beta_j} \log (1- \sigma (u))\\
   = \sum_{i=1}^M y^{(i)} (1-\sigma(u))x_j^{(i)} - (1-y^{(i)}) \sigma(u)x_j^{(i)} \label{eq:comp}
\end{eqnarray}


\begin{eqnarray}
   \frac{\partial \mathcal{L}}{\partial \beta_j} &=& \sum_{i=1}^M [y^{(i)} - \sigma(\beta^T\hat{x}^{(i)})]\hat{x_j}^{(i)}  \\
   &=& \sum_{i=1}^M [y^{(i)} -\hat{y}^{(i)}]\hat{x_j}^{(i)}
\end{eqnarray}

Intuitively, the above equation states that the likelihood gradient with respect to $\beta_j$ equals to the error term $y^{(i)} -\hat{y}^{(i)}$ multiplied by the datapoint's feature $\hat{x_j}^{(i)}$ that the parameter is responsible for, summed over all datapoints in the dataset.
Denoting the new parameter vector $\hat{\beta}$, the parameter update procedure can therefore be written as:

\begin{equation}
    \hat{\beta} = \beta + \eta \frac{\partial }{\partial \beta} \mathcal{L}(\beta)
\end{equation}

The parameter update can be done incorporating the vectorized format of equation, using the dataset matrices $\mathcal{X}$ and $\mathcal{Y}$:
\begin{equation}
    \hat{\beta} = \beta + \eta \sum_{i=1}^M [y^{(i)} -\sigma(\beta^T\hat{x}^{(i)})]\hat{x}^{(i)} \\
    \hat{\beta} = \beta + \eta \mathcal{X}^T(\mathcal{Y}-\sigma(\mathcal{X}\beta))
\end{equation}



## Import

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

## Logistic Regression functions

```python
def logistic(x): return 1.0 / (1.0 + np.exp(-x))

def pad_ones(X):
    return np.append(np.ones((X.shape[0],1)), X, axis = 1)

def logistic_regression(X_data, Y_data, learning_rate, max_it ):
    
    loss = []
    X_data = pad_ones(X_data)
    beta = np.random.uniform(-1, 1, X_data.shape[1])
    
    for i in range (0, max_it):
        y_predict = logistic(np.dot(X_data,beta))
        
        loss.append(compute_average_ll(X_data, Y_data, beta))
        
        beta = beta + learning_rate * np.dot(np.transpose(X_data) ,(Y_data - y_predict))
        
    return beta, loss

##
# X: 2d array with the input features
# y: 1d array with the class labels (0 or 1)
# w: current parameter values
#

def compute_average_ll(X, y, w):
    output_prob = logistic(np.dot(X, w))
    return - np.mean(y * np.log(output_prob) + (1 - y) * np.log(1.0 - output_prob))

##
# ll: 1d array with the average likelihood per data point and dimension equal
#     to the number of training epochs.
#

def pred_y_thresh(y_pred, thresh = 0.5):
    y_pred[y_pred > thresh] = 1
    y_pred[y_pred <= thresh] = 0
    return y_pred
```

## Data Visualisation Functions

```python
def plot_data_internal(X, y):
    x_min, x_max = X[ :, 0 ].min() - .5, X[ :, 0 ].max() + .5
    y_min, y_max = X[ :, 1 ].min() - .5, X[ :, 1 ].max() + .5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), \
        np.linspace(y_min, y_max, 100))
    plt.figure(figsize=(10,10))
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    ax = plt.gca()
    ax.plot(X[y == 0, 0], X[y == 0, 1], 'ro', markersize=4, label = 'Class 1')
    ax.plot(X[y == 1, 0], X[y == 1, 1], 'bo', markersize=4, label = 'Class 2')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Plot data')
    plt.legend(loc = 'upper left', scatterpoints = 1, numpoints = 1)
    return xx, yy

##
# X: 2d array with the input features
# y: 1d array with the class labels (0 or 1)
#

def plot_data(X, y):
    xx, yy = plot_data_internal(X, y)
    plt.show()
```

```python
#plotting the average loss likelihood against steps
def plot_ll(ll, title):
    plt.figure()
    ax = plt.gca()
    plt.xlim(0, len(ll) + 2)
    plt.ylim(min(ll) - 0.1, max(ll) + 0.1)
    ax.plot(np.arange(1, len(ll) + 1), ll, 'r-')
    plt.xlabel('Steps')
    plt.ylabel(title)
    plt.title('Plot {} Curve'.format(title))
    plt.show()
```

```python
def plot_predictive_distribution(X, y, w, predict):
    xx, yy = plot_data_internal(X, y)
    ax = plt.gca()
    X_predict = np.concatenate((xx.ravel().reshape((-1, 1)), \
        yy.ravel().reshape((-1, 1))), 1)
    Z = predict(X_predict,w)
    Z = Z.reshape(xx.shape)
    cs2 = ax.contour(xx, yy, Z, cmap = 'RdBu', linewidths = 3)
    plt.clabel(cs2, fmt = '%2.1f', colors = 'k', fontsize = 14)
    plt.show()
    
def predict_for_plot(x,w): 
    x_tilde = np.concatenate((np.ones((x.shape[ 0 ], 1 )), x), 1)
    return logistic(np.dot(x_tilde, w))
```

## Loading Data

```python
X = np.loadtxt('X.txt')
y = np.loadtxt('y.txt')

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 69)
```

## Visualising Data

```python
plot_data(X_train, Y_train)
print("X_train has shape: {}, y_train has shape: {}".format(X_train.shape, Y_train.shape))
```

## Logistic Regression - find the coefficients $\beta$

```python
beta, loss = logistic_regression(X_train ,Y_train , 0.003 , 100)
```

```python
plot_ll(loss, 'Average Training log-likelihood')
print("Final training log-likelihood: {}".format(loss[-1]))
```

```python
plot_predictive_distribution(X_train, Y_train, beta, predict_for_plot)
```

```python
ll_train = compute_average_ll(pad_ones(X_train), Y_train, beta)
ll_test = compute_average_ll(pad_ones(X_test), Y_test, beta)
```

```python
print("Final training log-likelihood: {}".format(ll_train))
print("Final test log-likelihood: {}".format(ll_test))
```

## Model Validation

### Confusion Matrix

```python
Y_pred = pred_y_thresh(logistic(np.dot(pad_ones(X_test), beta)),0.5)
c = confusion_matrix(Y_test, Y_pred)
c_pred_norm = c/np.tile([c[:,0].sum(),c[:,1].sum()],(2,1))
f1 = f1_score(Y_test, Y_pred, average='binary')


print("Test Set Confusion Matrix: \n{}".format(c_pred_norm))
print("f1_score: {}".format(f1))
```

# Logistic Regression with Gaussian Radial Basis Functions (RBFs)

Non-linearities are introduced to the input features by expanding the input through a set of gaussian radial basis functions (RBFs) centred on the training datapoints. 

$$ \hat{x}^{(n)}_{m+1} = exp\Bigg(-\frac{1}{2l^2}\sum_{d=1}^2(x_d^{(n)}-x_d^{(m)})^2\Bigg) $$


```python
##
# l: hyper-parameter for the width of the Gaussian basis functions
# Z: location of the Gaussian basis functions
# X: points at which to evaluate the basis functions

def expand_inputs(l, X, Z):
    X2 = np.sum(X**2, 1) 
    Z2 = np.sum(Z**2, 1)
    ones_Z = np.ones(Z.shape[ 0 ]) #no. of X_train data points (=800)
    ones_X = np.ones(X.shape[ 0 ])
    r2 = np.outer(X2, ones_Z) - 2 * np.dot(X, Z.T) + np.outer(ones_X, Z2)
    return np.exp(-0.5 / l**2 * r2)

##
# x: 2d array with input features at which to compute the predictions
#    using the feature expansion
#
# (uses parameter vector w and the 2d array X with the centers of the basis
# functions for the feature expansion, which are defined outside the function's
# scope)
#

def predict_for_plot_expanded_features(x,w):
    x_expanded = expand_inputs(rbf_kernel_size, x, X_train)
    x_tilde = np.concatenate((np.ones((x_expanded.shape[ 0 ], 1 )), x_expanded), 1)
    return logistic(np.dot(x_tilde, w))
```

```python
rbf_kernel_size = 0.01
X_train_rbf = expand_inputs(rbf_kernel_size, X_train, X_train)
X_test_rbf = expand_inputs(rbf_kernel_size, X_test, X_train)
```

```python
beta_rbf, loss_rbf = logistic_regression(X_train_rbf, Y_train, 0.003, max_it = 3000)
plot_ll(loss_rbf, 'Average Training log-likelihood')
print("Final training log-likelihood: {}".format(loss_rbf[-1]))
```

```python
plot_predictive_distribution(X_train, Y_train, beta_rbf, predict_for_plot_expanded_features)
```

```python
X_test_rbf = expand_inputs(rbf_kernel_size, X_test, X_train)
Y_pred = pred_y_thresh(logistic(np.dot(pad_ones(X_test_rbf), beta_rbf)),0.5)
c = confusion_matrix(Y_test, Y_pred)
c_pred_norm = c/np.tile([c[:,0].sum(),c[:,1].sum()],(2,1))
print("Test set confusion matrix: \n{}, kernel size is {}".format(c_pred_norm, rbf_kernel_size))
```

```python
rbf_sizes = [0.01, 0.1, 1]
for rbf_kernel_size in rbf_sizes:
    X_train_rbf = expand_inputs(rbf_kernel_size, X_train, X_train)
    X_test_rbf = expand_inputs(rbf_kernel_size, X_test, X_train)
    beta_rbf, loss_rbf = logistic_regression(X_train_rbf, Y_train, 0.003, max_it = 3000)
    print("RBF Kernel Size: {}".format(rbf_kernel_size))
    
    Y_pred = pred_y_thresh(logistic(np.dot(pad_ones(X_test_rbf), beta_rbf)),0.5)
    c = confusion_matrix(Y_test, Y_pred)
    c_pred_norm = c/np.tile([c[:,0].sum(),c[:,1].sum()],(2,1))
    print("Test set confusion matrix: \n{}".format(c_pred_norm))
    
    ll_train_rbf = compute_average_ll(pad_ones(X_train_rbf), Y_train, beta_rbf)
    ll_test_rbf = compute_average_ll(pad_ones(X_test_rbf), Y_test, beta_rbf)
    print("Final training log-likelihood: {}".format(ll_train_rbf))
    print("Final test log-likelihood: {}".format(ll_test_rbf))
    
    plot_predictive_distribution(X_train, Y_train, beta_rbf, predict_for_plot_expanded_features)
```
